# Notebook Purpose

This Jupyter notebook fetches historical commodities market data for any commodities using the yfinance library and saves it as a clean CSV file. It provides configurable global variables for the ticker symbol, date range, and data frequency (daily, hourly, etc.) to easily adapt the download to different commodities or periods. The output CSV includes Date, Open, High, Low, Close, Adjusted Close, and Volume columns with no extraneous header rows.

| Gold          | GC=F   | Gold Futures   |
| Silver        | SI=F   | Silver Futures |
| Crude Oil WTI | CL=F   | Crude Oil WTI  |

| Wheat         | ZW=F   | Wheat          |
| Corn          | ZC=F   | Corn           |
| Soybeans      | ZS=F   | Soybeans       |

```python
TICKERS = ["GC=F", "SI=F"]
START_DATE = "2010-04-01"
END_DATE = "2026-03-01"
INTERVAL = "1d"
DATA_DIR = "data/test_mean_reversion"
```

```python
import os
import yfinance as yf
import pandas as pd

os.makedirs(DATA_DIR, exist_ok=True)

for TICKER in TICKERS:
    df = yf.download(
        TICKER,
        start=START_DATE,
        end=END_DATE,
        interval=INTERVAL,
        auto_adjust=False,
        progress=False,
    )
    
    # Clean format
    df.columns = df.columns.droplevel(1)
    df.reset_index(inplace=True)
    
    csv_filename = f"{TICKER}_{INTERVAL}_{START_DATE}_{END_DATE}.csv"
    csv_path = os.path.join(DATA_DIR, csv_filename)
    df.to_csv(csv_path, index=False)
    
    print(f"✓ {TICKER}: {len(df)} rows → {csv_filename}")
```
