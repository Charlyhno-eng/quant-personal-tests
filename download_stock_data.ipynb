# Notebook Purpose

This Jupyter notebook fetches historical stock market data for any stock using the yfinance library and saves it as a clean CSV file. It provides configurable global variables for the ticker symbol, date range, and data frequency (daily, hourly, etc.) to easily adapt the download to different stocks or periods. The output CSV includes Date, Open, High, Low, Close, Adjusted Close, and Volume columns with no extraneous header rows.

```python
TICKER = "AAPL"
START_DATE = "2017-01-01"
END_DATE = "2026-01-01"
INTERVAL = "1d"  # ex: "1d", "1h", "15m"
DATA_DIR = "data/stocks_1d"
```

```python
import os
import yfinance as yf
import pandas as pd

os.makedirs(DATA_DIR, exist_ok=True)

df = yf.download(
    TICKER,
    start=START_DATE,
    end=END_DATE,
    interval=INTERVAL,
    auto_adjust=False,
    progress=False,
)

# Clean cleaning for 1 ticker only
df.columns = df.columns.droplevel(1)  # Delete an unnecessary line
df.reset_index(inplace=True)          # Date becomes column

# Naming with dates
csv_filename = f"{TICKER}_{INTERVAL}_{START_DATE}_{END_DATE}.csv"
csv_path = os.path.join(DATA_DIR, csv_filename)
df.to_csv(csv_path, index=False)

print(f"Données sauvegardées dans: {csv_path}")
print(df.head())
print(f"Shape: {df.shape}")
```
